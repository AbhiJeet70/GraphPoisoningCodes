```python
# Install necessary packages
!pip install torch_geometric
!pip install ogb

# Import necessary libraries
from torch_geometric.nn import GCNConv
from torch_geometric.datasets import Planetoid, Flickr
from torch_geometric.transforms import NormalizeFeatures
from ogb.nodeproppred import PygNodePropPredDataset
import torch
import torch.nn.functional as F
from torch.nn import Linear
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Helper function for visualizing embeddings
def visualize_emb(h, color, title):
    z = TSNE(n_components=2).fit_transform(h.detach().cpu().numpy())
    plt.figure(figsize=(10, 10))
    plt.xticks([])
    plt.yticks([])
    plt.scatter(z[:, 0], z[:, 1], s=70, c=color, cmap="Set2")
    plt.title(title)
    plt.show()

# Load datasets function
def load_dataset(dataset_name):
    """
    Load the specified dataset for training.
    """
    if dataset_name in ["Cora", "PubMed"]:
        dataset = Planetoid(root=f"./data/{dataset_name}", name=dataset_name, transform=NormalizeFeatures())
    elif dataset_name == "Flickr":
        dataset = Flickr(root="./data/Flickr")
    elif dataset_name == "OGB-arxiv":
        dataset = PygNodePropPredDataset(name="ogbn-arxiv", root="./data/OGB-arxiv")
    else:
        raise ValueError(f"Unsupported dataset: {dataset_name}")
    return dataset

# Define the MLP model
class MLP(torch.nn.Module):
    def __init__(self, hidden_channels, num_features, num_classes):
        super().__init__()
        torch.manual_seed(12345)
        self.lin1 = Linear(num_features, hidden_channels)
        self.lin2 = Linear(hidden_channels, num_classes)

    def forward(self, x):
        x = self.lin1(x).relu()
        x = F.dropout(x, p=0.5, training=self.training)
        return self.lin2(x)

# Define the GCN model
class GCN(torch.nn.Module):
    def __init__(self, hidden_channels, num_features, num_classes):
        super().__init__()
        torch.manual_seed(1234567)
        self.conv1 = GCNConv(num_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, num_classes)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index).relu()
        x = F.dropout(x, p=0.5, training=self.training)
        return self.conv2(x, edge_index)

# Function to train and visualize embeddings for each dataset
def train_and_visualize(dataset_name, model_type="MLP"):
    # Load dataset
    dataset = load_dataset(dataset_name)
    data = dataset[0]

    # Choose the model type
    if model_type == "MLP":
        model = MLP(hidden_channels=16, num_features=dataset.num_features, num_classes=dataset.num_classes)
    elif model_type == "GCN":
        model = GCN(hidden_channels=16, num_features=dataset.num_features, num_classes=dataset.num_classes)
    else:
        raise ValueError(f"Unsupported model type: {model_type}")

    # Define loss and optimizer
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

    # Train the model
    def train():
        model.train()
        optimizer.zero_grad()
        if model_type == "MLP":
            out = model(data.x)
        elif model_type == "GCN":
            out = model(data.x, data.edge_index)
        loss = criterion(out[data.train_mask], data.y[data.train_mask])
        loss.backward()
        optimizer.step()
        return loss

    # Run training
    for epoch in range(1, 201):
        loss = train()
        if epoch % 10 == 0:
            print(f'Dataset: {dataset_name}, Epoch {epoch}, Loss: {loss:.4f}')

    # Evaluate and visualize embeddings
    model.eval()
    if model_type == "MLP":
        out = model(data.x)
    elif model_type == "GCN":
        out = model(data.x, data.edge_index)

    # Visualize embeddings with separate plots for each dataset
    visualize_emb(out, color=data.y, title=f"Clustering for {dataset_name} Dataset using {model_type}")

# Run for each dataset
dataset_names = ["Cora", "PubMed", "Flickr"]
for name in dataset_names:
    print(f"\nTraining and visualizing for {name} with MLP model")
    train_and_visualize(name, model_type="MLP")

    print(f"\nTraining and visualizing for {name} with GCN model")
    train_and_visualize(name, model_type="GCN")
```
